```
df = pd.read_csv("/content/kirby21.csv")
```

```
df.head(4)
```

```
df = df.drop(['Unnamed: 0', 'rawid', 'min', 'max'],\
             axis = 1)
```

```
## Extract the Type 1 Level 1 data
t1l1 = df.loc[(df.type == 1) & (df.level == 1)]
## Create a new column based on ICV
t1l1 = t1l1.assign(icv = sum(t1l1.volume))
t1l1
```

```
df[df['roi'] == 'Telencephalon_L']['volume']
xlist = df[df['roi'] == 'Telencephalon_L']['volume']

df[df['roi'] == 'Telencephalon_R']['volume']
ylist = df[df['roi'] == 'Telencephalon_R']['volume']
```

```
import matplotlib.pyplot as plt

plt.scatter(xlist, ylist, c='b', alpha = 0.6)
plt.xlabel('Telencephalon_L volume')
plt.ylabel('Telencephalon_R volume')
```

```
x_mean = df[df['roi'] == 'Telencephalon_L']['mean']
y_mean = df[df['roi'] == 'Telencephalon_R']['mean']

x_new = xlist - x_mean
y_new = ylist - y_mean

plt.scatter(x_new, y_new, c='g', alpha = 0.6)
plt.xlabel('Telencephalon_L demean volume')
plt.ylabel('Telencephalon_R demean volume')

from pylab import * 

m,b = polyfit(x_new, y_new, 1) 

plot(x_new, y_new, 'yo', x_new, m*x_new+b, '--k') 
plt.show() 
print('slope: ', m)
```
